# Circuits

## Bell state

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.draw("mpl")
```

## Deutsch's algorithm

```python
qc = QuantumCircuit(2, 1)
qc.x(1)
qc.h(0)
qc.h(1)
qc.barrier()
# UNITARY here?
qc.barrier()
qc.h(0)
qc.measure([0], [0])
qc.draw("mpl")
```
