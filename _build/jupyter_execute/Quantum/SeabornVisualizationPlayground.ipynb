# Imports

```python
import seaborn as sns
```

# Jointplot

```python
penguins = sns.load_dataset("penguins")
sns.jointplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    kind="reg",
)
```

```python
sns.jointplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species"
)
```

```python
sns.jointplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    kind="reg"
)
```
