```python
def generate_plot(
        conversions: Dict[str, float],
        x_axis_label: str = ""
) -> None:
    data = [(k, v) for k, v in sorted(conversions.items(), key=lambda x: -x[1])]
    plt.figure(figsize=(10, .4 * len(data)))
    plt.barh(
        [x[0] for x in data],
        width=[x[1] for x in data],
        height=.8,
        log=True,

    )
    plt.xlabel(x_axis_label)
```

```python
generate_plot(
    x_axis_label="Joules ( SI )",
    conversions={
        "calorie": 4184,
        "hartree": 4.3597e-18,
        "joule": 1,
        "kelvin": 1.380648780669e-23,
        "MeV": 1.60218e-16,
        "GeV": 1.60218e-13,
        "TeV": 1.60218e-10,
        "PeV": 1.60218e-7,
        "ZeV": 1.60218e-1,
        "eV": 1.60218e-19,
        "kWh": 3.6e+6,
        "megajoules": 10e6,
    }
)
```
