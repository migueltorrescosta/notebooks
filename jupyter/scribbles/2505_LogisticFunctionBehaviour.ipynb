# LogisticFunctionBehaviour

```python
from matplotlib import pyplot as plt
```

```python
def f(r, x):
    return r * x * (1 - x)


def sample(x0, r, n):
    v = [x0]
    for _ in range(n):
        v.append(f(r, v[-1]))
    plt.figure(figsize=(30, 3))
    plt.plot(v)
```

```python
x0 = 0.000001
r = 3.87
n = 100
sample(x0, r, n)
```
