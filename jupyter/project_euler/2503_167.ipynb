```python
def u(a, b, k):
    if k == 1:
        return {a}
    elif k == 2:
        return {a, b}
    previous_members = u(a, b, k - 1)
    sums = Counter(
        [
            i + j
            for (i, j) in itertools.product(previous_members, previous_members)
            if i < j
        ]
    )
    least_unique = min(
        {
            number
            for number, count in sums.items()
            if count == 1 and number > max(previous_members)
        }
    )
    return u(a, b, k - 1).union({least_unique})
```
