```python
def sample_points(n):
    x_coordinates = list(map(lambda x:2*x-1,np.random.rand(n)))
    y_coordinates = list(map(lambda x:2*x-1,np.random.rand(n)))
    points = list(zip(x_coordinates,y_coordinates))
    return points
```

```python
sample_points(100)
```
