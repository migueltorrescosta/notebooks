# Circuits

## Bell state

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.draw("mpl")
```

## Quantum teleportation

The below does not seem sufficient as it does not show measurements :(

```python
qc = QuantumCircuit(3, 2)

qc.h(2)
qc.sdg(2)
# qc.save_statevector('$\\psi_1$')

qc.h(1)
qc.cx(1, 0)
qc.barrier()

qc.cx(2, 1)
qc.h(2)
qc.barrier()

qc.measure([2, 1], [1, 0])
qc.z(0).c_if(1, 1)
qc.x(0).c_if(0, 1)
# qc.save_statevector('$\\psi_2$')

qc.draw("mpl")
```
