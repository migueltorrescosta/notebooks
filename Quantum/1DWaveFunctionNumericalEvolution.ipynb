# Goal

We have a potential which has different eigenstates $|i>$. We want to devise a control potential $V$ that evolves the state $|0>$ to $|4>$ without losing any information to other states. In practice we want $<0|V|n>=1$, and all other transitions to be zero. This should be achievable by having $V = \frac{|n>}{|0>}$

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
xl, xr =0,4 #declaring left/right side potentials barrier
L = xr-xl
xx=np.linspace(xl,xr,1000) #points where ψ and ψ*ψ are to be evaluated

potential = [np.cos(10*x)+(x-2)**4/2 for x in xx]


A=pow(2/L,0.5) # Normalization constant
def wave_function_generator(x: list[float], n:float) -> list[float]:
    return A*np.sin(n*np.pi*x/L)

control_potential = np.divide(
    wave_function_generator(xx,5),
    wave_function_generator(xx,0)
)

plt.figure(figsize=(16,6))
plt.plot(xx,potential,label="potential")
plt.plot(xx,potential,label="control_potential")
for i in range(0,3):
    plt.plot(xx, wave_function_generator(xx, i), label=f"$\psi_{i}$(x)")
plt.xlabel('x')
plt.title('State')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),fancybox=True, shadow=True, ncol=10)
plt.grid()
```
